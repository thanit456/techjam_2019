```python
demograph_df = pd.read_csv('demographics.csv')
cc_df = pd.read_csv('cc.csv')
kplus_df = pd.read_csv('kplus.csv')
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
cc_df['pos_dt'][0]
```

```python
def str2date(s):
    formatter = '%Y-%m-%d'
    return datetime.strptime(s, formatter)

cc_df['pos_dt'] = cc_df['pos_dt'].apply(lambda x: str2date(x))
kplus_df['sunday'] = kplus_df['sunday'].apply(lambda x: str2date(x))
```

```python
train_demograph_df = pd.merge(demograph_df, train_df, on='id', how= 'inner')
train_demograph_df
```

```python
train_demograph_df = train_demograph_df.dropna()
```

```python
train_kplus_df = pd.merge(train_demograph_df, kplus_df, on='id', how='inner')
train_kplus_df
```

```python
sunday_df = pd.DataFrame()
sunday_df['cc_no'] = train_kplus_df['cc_no']
sunday_df['sunday'] = train_kplus_df['sunday']
sunday_df
```

```python
sunday_cc_df = pd.merge(sunday_df, cc_df, on='cc_no', how='inner')
sunday_cc_df
```

```python
sunday_cc_df['less7'] = (sunday_cc_df['sunday'] - sunday_cc_df['pos_dt'] < datet.timedelta(7)) 
sunday_cc_df['emore0'] = (sunday_cc_df['sunday'] - sunday_cc_df['pos_dt'] >= datet.timedelta(0)) 
```

```python
sunday_cc_df = sunday_cc_df[sunday_cc_df['less7'] == True]
sunday_cc_df = sunday_cc_df[sunday_cc_df['emore0'] == True]
```

```python
sunday_cc_df = sunday_cc_df.drop('less7', axis=1)
sunday_cc_df = sunday_cc_df.drop('emore0', axis=1)
```

```python
def get_fc_model():
    
    model = Sequential([
        Dense(100, input_shape=(3,) ,activation='relu'),
        Dropout(0.5),
        Dense(50, activation='relu'),
        Dense(1)
    ])
    
    model.compile(optimizer=Adam(learning_rate=lr, decay=decay),loss='mse', metrics=['mse'])
    
    model.summary()
    return model
```
